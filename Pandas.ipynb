2.Dataframe:Pandas DataFrames are two-dimensional data structures with labeled rows and columns, that can hold many data types.

### Sorting and subsetting

```python
df.sort_index() #sort the dataframe through the index.
```

```python
df.sort_values('Industry_code_NZSIOC',ascending=True)
```

```python
df.sort_values(['Industry_code_NZSIOC','Year'],ascending=False) #sorting by multiple values
```

```python
df.sort_values('Year',ascending=True)
```

```python
df.sort_values(['Industry_code_NZSIOC','Year'],ascending=[False,True]) #sort first column by descending order and Year by ascending order.
```

```python
#Subsetting columns
df['Year']
```

```python
df.Year
```

```python
columns_selected=['Year','Units']
df[columns_selected]
```

```python
df[['Year','Units']]
```

```python
#Subsetting rows
df[df['Year']>2018]
```

```python
df[df['Units']=='Percentage']
```

```python
df['Year'][df['Units']=='Dollars (millions)']
```

```python
df[['Year','Industry_code_NZSIOC','Units']][df['Units']=='Dollars (millions)']
```

```python
df[(df['Year']>2018)& (df['Units']=='Percentage')]
```

```python
df[(df['Year']>2018)| (df['Units']=='Percentage')] #Logical OR
```

```python
#Subsetting using isin(),categorical values
year_2019=df['Year'].isin([2019])
year_2019
```

```python
df[year_2019]
```

```python
df[df['Industry_aggregation_NZSIOC'].isin(['Level1','Level 3'])]
```
